```python
# Spatial Join?
joined = gpd.sjoin(wheeling_par_within, wheeling_holc)
joined
```

```python
joined.plot()
```

```python
joined['holc_id'].unique()
```

```python
joined['holc_grade'].unique()
```

```python
# For each 'holc_id' get the stats of Residential single family homes
joined[(joined['ExmptCd'] == 'Residential') & (joined['holc_id'] == 'A1')]
```

```python
def get_sum(df, holc_id, col):
    temp = df[(df['ExmptCd'] == 'Residential') & (df['holc_id'] == holc_id)]
    return temp[col].sum()
```

```python
get_sum(joined, 'A1', 'Acre')
```

```python
columns = [
    'Area Code',
    'Parcel Count',
    'Acres',
    'Total Land Value',
    'Total Building Value',
    'Total Market Value',
    'Total Taxable Land Value',
    'Total Taxable Building Value',
    'Total Taxable Value',
    'Average Market VPA',
    'Average Taxable VPA',
    'GIS Avg Market VPA',
    'GIS VPA Median',
    'GIS Grade Median VPA',
    'GIS Grade Average VPA',
    'VPA Check',
    'By Color'
]
```

```python
df2 = pd.DataFrame(columns = columns)
df2
```

```python
df2['Area Code'] = joined['holc_id'].unique()
```

```python
df2
```

```python
df2 = df2.sort_values(by = 'Area Code', ignore_index = True)
df2
```

```python
df2[df2['Area Code'] == joined['holc_id'].unique()[0]]['Acres'] = 1
```

```python
res_joined = joined[joined['ExmptCd'] == 'Residential']
res_joined = res_joined.reset_index(drop=True)
res_joined
```

```python
res_joined['LndVal'] = res_joined['LndVal'].astype('int64')
res_joined['BdVal'] = res_joined['BdVal'].astype('int64')
```

```python
assessment_ratio = 0.6
```

```python
for i in df2['Area Code']:
    # Number of parcels in each Area Code
    df2.loc[df2['Area Code'] == i, 'Parcel Count'] = len(res_joined[res_joined['holc_id'] == i])
    # Sum of Acres in each Area Code
    df2.loc[df2['Area Code'] == i, 'Acres'] = get_sum(res_joined, i, 'Acre')
    # Total Land Value
    df2.loc[df2['Area Code'] == i, 'Total Land Value'] = get_sum(res_joined, i, 'LndVal')
    # Total Building Value
    df2.loc[df2['Area Code'] == i, 'Total Building Value'] = get_sum(res_joined, i, 'BdVal')
    # Total Market Value
    df2.loc[df2['Area Code'] == i, 'Total Market Value'] = get_sum(res_joined, i, 'MrkVal')
    # Average Market VPA
    df2.loc[df2['Area Code'] == i, 'Average Market VPA'] = res_joined[res_joined['holc_id'] == i]['VPA'].mean()
#     df2.loc[df2['Area Code'] == i, 'Total Taxable Land Value'] = df2['Total Land Value'] * assessment_ratio
df2['Total Taxable Land Value'] = df2['Total Land Value'] * assessment_ratio 
df2['Total Taxable Building Value'] = df2['Total Building Value'] * assessment_ratio
df2['Total Taxable Value'] = df2['Total Market Value'] * assessment_ratio
df2['Average Taxable VPA'] = df2['Total Taxable Value'] / df2['Acres']
df2
```

```python
df2.to_csv("Wheeling_redlining_values.csv")
```
